### 01. DataFrame 인덱싱
- 데이터 분석의 덕목 : 자유자재 인덱싱
    - 손에 익을 때까지 무한 연습!!
---
###### *df : 임시 데이터프레임 변수 이름*
###### 핵심: 간단한 코드 -> 굳이 안써도 되는건 쓰지 말자! (ex. loc)
---
- 하나의 값을 받아오자.:: DataFrame 기본 인덱싱 형태
    - df.loc['로우이름', '컬럼이름']
- 한 행을 통째로 받아오자.
    - df.loc['로우이름', :] 
    - df.loc['로우이름'] 
        - ':'는 처음부터 끝까지를 의미
- 한 열을 통째로 받아오자.
    - df['열이름']
    - df.loc[:, '열이름']
- 여러 개의 열을 통째로 받아오자. -> 리스트 인덱싱
    - df[['A열 이름', 'B열 이름', ...]] 
        - 대괄호 두개 : df[] 인덱싱에 하나의 자료만 들어가야하는 점! -> 여기서는 '리스트'형 자료를 넣은 것.
        - 열은 loc 꼭 안붙여도 ㄱㅊ
    - df.loc[:, ['A열 이름', 'B열 이름', ...]]
- 여러 개의 행을 통째로 받아오자. -> 리스트 인덱싱
    - df.loc[['x행 이름', 'y행 이름', ...]]
---
###### 주의
- 행/열 이름이 숫자이면 ' ' 없이 그냥 바로 쓰자.

### 06. DataFrame 인덱싱2
#### 슬라이싱
- 행 슬라이싱
    - df.loc['시작 행 이름': '끝 행 이름']
---
#### 주의 :: Column 슬라이싱
- 열 슬라이싱
    - df.loc[:, '시작 열 이름':'끝 열 이름']
        - : -> 모든 row에 대해서..!
- 행&열 모두 포함 슬라이싱
    - df.loc['시작행 이름':'끝 행 이름', '시작 열 이름':'끝 열 이름']

### 08. Data Frame 조건으로 인덱싱
cf. 불린값으로 받아오기  

--- 

- df['열 이름'] >/</= 조건
##### 필터링(인덱싱)
- df.loc[df['열 이름'] >/</== 조건]
    - 해당 조건에 맞는 요소만 필터링 됨
- 여러 조건을 만족시키는 요소를 찾고 싶을 때
    - (조건) &/| (조건)
    - 그걸 다시 필터링 ~

### 11. DataFrame 위치로 인덱싱하기
(numpy array 처럼 인덱스(위치)로 인덱싱하기)  
##### df.iloc

- df.iloc[행 인덱스, 열 인덱스]
- 여러값을 동시에 받아오기
    - df.iloc[[행,열],[행,열]]
- 슬라이싱
    - df.iloc[3:, 1:4] :: 3행, 1~3열
