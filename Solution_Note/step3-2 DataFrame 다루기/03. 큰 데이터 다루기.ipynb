### 01. 큰 DataFrame 살펴보기
데이터가 방대한 경우, 이를 보기좋게 컨트롤하는 능력 필요

#### 데이터 프레임 확인
- 데이터 프레임의 전체 스케일 
    - df.shape
- 콜럼 확인
    - df.columns
- 각 콜럼에 대한 기본 정보
    - df.info()
- DF의 통계 정보
    - df.describe() : 각 컬럼에 대한 통계값들(편차 등등)

    
- 앞의 n줄만 보고 싶을 때
    - df.head(n)
- 뒤에 n줄만 보고 싶을 때
    - df.tail(n)

---
#### 데이터 프레임 정렬 
- df.sort_values(by='price')
- 내림차순 : ascending = False
    - df.sort_values(by='price', ascending = False)
- 본질적인 수정 X -> inplace = True해야 바뀜 
    - df.sort_values(by='price', ascending = False, inplace = True)

    

```python
import pandas as pd
```

```python
laptops_df = pd.read_csv('data/laptops.csv')
laptops_df
```

```python
laptops_df.sort_values(by = 'price', ascending = True, inplace = True)
laptops_df
```

### 02. 큰 Series 살펴보기
예제로 확인

```python
print('어떤 컬럼(브랜드)만 보고 싶을 때')
print(laptops_df['brand'])

print("=========================")
print("=========================")

# 중복 제거해서 보기..
print('컬럼 값들이 겹치는 경우, 중복 제거하고 싶을 때')
print(laptops_df['brand'].unique())

print("=========================")
print("=========================")

# 컬럼 각 요소 접근
print('각 컬럼 값에서 값들의 개수를 세고 싶을 때')
print(laptops_df['brand'].value_counts())

print("=========================")
print("=========================")

# 컬럼 값들의 요약본 
print(laptops_df['brand'].describe()) # 전체 개수, 중복 제외한 값, 최다 빈출 값, 
```
